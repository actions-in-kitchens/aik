```python
%matplotlib inline
import matplotlib.pyplot as plt
import json
import cv2
from os.path import isdir, join
Settings = json.load(open('settings.json'))
dataset_dir = Settings['dataset']
n_cameras = Settings['n_cameras']
assert isdir(dataset_dir), dataset_dir
vid_dir = join(dataset_dir, 'videos')
assert isdir(vid_dir), vid_dir
cam_dir = join(dataset_dir, 'cameras')
assert isdir(cam_dir), cam_dir
FRAME = 1
I = []
cams = []
for cid in range(n_cameras):
    local_vid_dir = join(vid_dir, 'camera%02d' % cid)
    local_cam_dir = join(cam_dir, 'camera%02d' % cid)
    im_file = join(local_vid_dir, 'frame%09d.png' % FRAME)
    cam_file = join(local_cam_dir, 'frame%09d.json' % FRAME)
    I.append(cv2.cvtColor(cv2.imread(im_file), cv2.COLOR_BGR2RGB))
    cam = json.load(open(cam_file))
    cams.append(cam)
```

```python
selected_cids = [0, 1, 2, 3]

fig = plt.figure(figsize=(16, 9))

for i, cid in enumerate(selected_cids):
    ax = fig.add_subplot(2, 2, i + 1)
    ax.axis('off')
    ax.imshow(I[cid])

plt.tight_layout()
plt.show()
```
